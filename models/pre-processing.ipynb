```python
import pandas as pd

df = pd.read_csv("India_Methane_Hotspots.csv")

df[["longitude", "latitude"]] = df[".geo"].str.extract(
    r'\[([\d\.\-]+),([\d\.\-]+)\]'
)

df["longitude"] = df["longitude"].astype(float)
df["latitude"] = df["latitude"].astype(float)

df.head()
```

```python
threshold1 = df["count"].quantile(0.50)
threshold2 = df["count"].quantile(0.85)

def classify(x):
    if x > threshold2:
        return "Severe"
    elif x > threshold1:
        return "Moderate"
    else:
        return "Low"

df["severity"] = df["count"].apply(classify)
```
